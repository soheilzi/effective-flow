```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np 
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15, 5)
# plt.figure(figsize=(8,6))
```

```python
t0 = lambda a, b, p : a * b if (p == .5) else ( (a * ((1-p)**(a + b)) + (b * (p**(a + b))) - ((a + b) * ((1-p)**a) * (p**b))) / ((1 - p*2) * (((1-p)**(a + b)) - (p**(a + b)))))
```

```python
alpha = 1.
ppts = np.linspace(0, 1, 100000)
for cap in [50, 60, 70]:
    a = cap / (1 + (1. / alpha))
    b = a / alpha
    ax = plt.plot(ppts, [t0(a, b, p) for p in ppts], label = "Capacity: %d"%cap)



plt.legend()
plt.title("Channel Capacity effect")
plt.xlabel("Probability")
plt.ylabel("Expected Life Span")
plt.show()
```

```python
cap = 50
ppts = np.linspace(0, 1, 100000)
for alpha in [1, 5, 50]:
    a = cap / (1 + (1. / alpha))
    b = a / alpha
    plt.plot(ppts, [t0(a, b, p) for p in ppts], label = "a/b: %d"%alpha)



plt.legend()
plt.title("Channel a/b effect")
plt.xlabel("Probability")
plt.ylabel("Expected Life Span")
plt.show()
```

```python
p = .5
cpts = np.linspace(0, 1000, 100000)
for alpha in [1, 5, 50]:
    plt.plot(cpts, [t0(cap / (1 + (1. / alpha)), (cap / (1 + (1. / alpha))) / alpha, p) for cap in cpts], label = "a/b: %d"%alpha)



plt.legend()
plt.title("Channel a/b effect")
plt.xlabel("Capacity")
plt.ylabel("Expected Life Span")
plt.show()
```

```python
## max life span per alpha
## max alpha = 10, min = 0.1
a = 10
ppts = np.linspace(0, 1, 1000)
maxLifeSpan = []
bList = np.linspace(a/10, 10 * a, 500)

for b in bList:
    maxLifeSpan.append(max([t0(a, b, p) for p in ppts]))

plt.plot(bList, maxLifeSpan, label = "a=%d"%10)

plt.legend()
plt.title("A=const effect")
plt.xlabel("Other Edge")
plt.ylabel("Maximum Expected Life Span")
plt.show()
```
